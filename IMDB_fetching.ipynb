```python
import pandas as pd
import requests
from time import sleep

# Your OMDb API Key
#OMDB_API_KEY = 'b9a1e312'  # Replace this with your actual OMDb API key

# Step 1: Load IMDb datasets
titles = pd.read_csv('title.basics.tsv.gz', sep='\t', na_values='\\N', low_memory=False)
ratings = pd.read_csv('title.ratings.tsv.gz', sep='\t', na_values='\\N')

# Merge datasets on 'tconst'
movies = pd.merge(titles, ratings, on='tconst')

# Filter movies with at least 50,000 ratings and type='movie'
movies_filtered = movies[
    (movies['titleType'] == 'movie') &
    (movies['numVotes'] >= 50000)
]

# Sort movies by rating descending and take top 1000
top_movies = movies_filtered.sort_values(by='averageRating', ascending=False).head(1000)

# OMDb API Key (only the key itself!)
OMDB_API_KEY = 'b9a1e312'
```

```python
# First convert the string representation of list to actual list using eval()
merged_df['Ratings'] = merged_df['Ratings'].apply(eval)

# Function to extract Rotten Tomatoes score and convert to 10-point scale
def extract_rt_score(ratings):
    try:
        for rating in ratings:
            if rating['Source'] == 'Rotten Tomatoes':
                # Remove the % sign and convert to float, then divide by 10
                return float(rating['Value'].strip('%')) / 10
    except:
        return None

# Convert Metascore to 10-point scale
merged_df['Metascore_10'] = merged_df['Metascore'].astype(float) / 10

# Extract and add Rotten Tomatoes score
merged_df['RottenTomatoes_10'] = merged_df['Ratings'].apply(extract_rt_score)

# Now you have three normalized scores:
# - imdbRating (already on 10-point scale)
# - Metascore_10 (converted to 10-point scale)
# - RottenTomatoes_10 (converted to 10-point scale)

# You can verify the results:
print("\nFirst few rows of the normalized scores:")
print(merged_df[['Title', 'imdbRating', 'Metascore_10', 'RottenTomatoes_10']].head())
```
